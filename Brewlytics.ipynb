```python
#changing the data types
brewery_data['Brew_Date']=pd.to_datetime(brewery_data['Brew_Date'])
brewery_data['Brew_Date1'] = brewery_data['Brew_Date'].dt.date
brewery_data['Brew_Date1']=pd.to_datetime(brewery_data['Brew_Date1'])
brewery_data['Brew_Time'] = brewery_data['Brew_Date'].dt.time
```

# Feature engineering

```python
brewery_data['Month_Year'] = brewery_data['Brew_Date'].dt.to_period('M').astype(str)
#Interaction Between Alcohol_Content and Gravity:
brewery_data['Interaction_Alcohol_Gravity']=brewery_data['Alcohol_Content']* brewery_data['Gravity']

#Interaction Between pH_Level and Bitterness
brewery_data['Inctraction_pH_Bitterness'] =brewery_data['pH_Level']* brewery_data['Bitterness']

#Interaction Between Temperature and Fermentation_Time
brewery_data['Interaction_Temperature_Fermentation'] = brewery_data['Temperature']* brewery_data['Fermentation_Time']
```
